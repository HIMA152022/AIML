```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bankcustdata = pd.read_csv('/content/Bank_customer_data.csv')

print(bankcustdata.info())
print(bankcustdata.describe())
print(bankcustdata.head())
```

```python
print(bankcustdata.isnull().sum())  # Count missing values in each column
```

```python
sns.countplot(x=bankcustdata["Personal Loan"])
plt.title("Loan Approval Distribution")
plt.show()

print(bankcustdata["Personal Loan"].value_counts(normalize=True))  # Check class percentages
```

```python
# Histogram of Numerical Features

bankcustdata.hist(figsize=(12, 12), bins=20, edgecolor="red")
plt.tight_layout()
plt.show()
```

```python
# Box Plots (Detect Outliers in Numerical Features)

plt.figure(figsize=(10, 5))
sns.boxplot(data=bankcustdata.drop(columns=["Personal Loan"]))  # Exclude target variable
plt.xticks(rotation=45)
plt.title("Box Plot of Numerical Features")
plt.show()
```

```python
# Density Plots (Understand Feature Distributions)

plt.figure(figsize=(10, 5))
for col in ["Income", "CCAvg", "Mortgage"]:
    sns.kdeplot(bankcustdata[col], label=col, fill=True)  # Use fill=True instead of shade=True
plt.title("Density Plots of Key Financial Features")
plt.legend()
plt.show()
```

```python
# Pair Plot (Explore Feature Relationships)

sns.pairplot(bankcustdata, hue="Personal Loan", diag_kind="kde")
plt.show()
```

```python
# Bar Chart for Categorical Feature

sns.countplot(x=bankcustdata["Education"], hue=bankcustdata["Personal Loan"])
plt.title("Education Level vs. Loan Approval")
plt.show()
```
