```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
dataset = pd.read_csv('/content/WineQT.csv')

print(dataset.head())
```

```python
# Check dataset dimensions
print(f"Dataset Shape: {dataset.shape}")

# Check column data types & missing values
print("\nData Info:")
print(dataset.info())

print("\nSummary Statistics:")
print(dataset.describe())

print("\nQuality Distribution:")  # wine quality counts
print(dataset["quality"].value_counts())
```

```python
# Plot bar chart for quality distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=dataset, x="quality", hue="quality", palette="viridis", legend=False)
plt.xlabel("Wine Quality")
plt.ylabel("Count")
plt.title("Distribution of Wine Quality")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=dataset, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Boxplot of Wine Features")
plt.show()
```

```python
# Alcohol vs Quality

plt.figure(figsize=(10, 5))
sns.scatterplot(x=dataset["alcohol"], y=dataset["quality"], hue=dataset["quality"], palette="coolwarm")
plt.xlabel("Alcohol Content")
plt.ylabel("Wine Quality")
plt.title("Alcohol vs Wine Quality")
plt.show()
```
