```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/tips.csv")

print(dataset.head())
```

```python
# Check data dimensions (rows & columns)
print(f"Data Shape: {dataset.shape}")
```

```python
# Check data types and missing values
print("\nData Info:")
print(dataset.info())
```

```python
# Check basic statistics for numerical columns
print("\nSummary basic Statistics:")
print(dataset.describe())
```

```python
# Scatter plot of total bill vs. tip
plt.figure(figsize=(8,6))
sns.scatterplot(x=dataset["total_bill"], y=dataset["tip"], hue=dataset["sex"])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Relationship Between Total Bill & Tip")
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = dataset.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".4f")
plt.title("Correlation matrices & Heatmaps to Explore Relationships vs Multiple features")
plt.show()
```

```python
# Box plot of tip amounts by day
plt.figure(figsize=(8,4))
sns.boxplot(x="day", y="tip", data=dataset)
plt.xlabel("Day of the Week")
plt.ylabel("Tip ($)")
plt.title("Tip Distribution by Day")
plt.show()
```

```python
# Histogram of tip amounts
plt.figure(figsize=(8,4))
sns.histplot(dataset["tip"], bins=10, kde=True)
plt.xlabel("Tip Amount ($)")
plt.ylabel("Frequency")
plt.title("Tip Amount Distribution")
plt.show()
```
